# Experiment analysis

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
alternative="exta_aid_180529"
nr_replications=10
dic30 = {}
dic0 = {}
for i in range(1,nr_replications+1):
    j=str(i)
    dic_key="rep"+j
    dic30[dic_key]=pd.read_excel("model_output_basicmodel_80bulk30_show30_Rep"+j+".xlsx")
    dic0[dic_key]=pd.read_excel("model_output_basicmodel_80bulk30_show0_Rep"+j+".xlsx")

```

```python
df_concat30 = pd.concat(dic30.values())
by_row_index30 = df_concat30.groupby(df_concat30.index)
df_means30 = by_row_index30.mean()

df_concat0 = pd.concat(dic0.values())
by_row_index0 = df_concat0.groupby(df_concat0.index)
df_means0 = by_row_index0.mean()
x=df_means30.shape[0]
```

```python
#PROCESSED
df_dock_processed=df_means30[["Time","dockselection_processed"]].merge(df_means0[["Time","dockselection_processed"]], left_on='Time', right_on='Time', how='outer')
df_dock_processed.rename(columns={'dockselection_processed_x': 'dockselection_processed30', 'dockselection_processed_y': 'dockselection_processed0'}, inplace=True)
plot_proc_dock=df_dock_processed.plot(title=alternative, x='Time', y=['dockselection_processed30','dockselection_processed0'], xticks=np.arange(0, x, 48.0))
plot_proc_dock.set_xlabel("hour of simulation")
plot_proc_dock.set_ylabel("cargo units")
fig=plot_proc_dock.get_figure()
fig.savefig("Results/Plots/processed_by_taxilane_"+alternative)
print(df_dock_processed.sum(axis=0))

df_unloading_processed=df_means30[["Time","unloading_processed"]].merge(df_means0[["Time","unloading_processed"]], left_on='Time', right_on='Time', how='outer')
df_unloading_processed.rename(columns={'unloading_processed_x': 'unloading_processed30', 'unloading_processed_y': 'unloading_processed0'}, inplace=True)
plot_proc_unload=df_unloading_processed.plot(title=alternative, x='Time', y=['unloading_processed30','unloading_processed0'], xticks=np.arange(0, x, 48.0))
plot_proc_unload.set_xlabel("hour of simulation")
plot_proc_unload.set_ylabel("cargo units")
fig=plot_proc_unload.get_figure()
fig.savefig("Results/Plots/processed_at_unloading_"+alternative)
print(df_unloading_processed.sum(axis=0))

df_warehouse_processed=df_means30[["Time","warehouse_processed"]].merge(df_means0[["Time","warehouse_processed"]], left_on='Time', right_on='Time', how='outer')
df_warehouse_processed.rename(columns={'warehouse_processed_x': 'warehouse_processed30', 'warehouse_processed_y': 'warehouse_processed0'}, inplace=True)
plot_proc_ware=df_warehouse_processed.plot(title=alternative, x='Time', y=['warehouse_processed30','warehouse_processed0'], xticks=np.arange(0, x, 48.0))
plot_proc_ware.set_xlabel("hour of simulation")
plot_proc_ware.set_ylabel("cargo units")
fig=plot_proc_ware.get_figure()
fig.savefig("Results/Plots/processed_at_warehouse_"+alternative)
print(df_warehouse_processed.sum(axis=0))
```

```python
#THROUGHPUT TIME

#taxi
df_Time_average_taxi=df_means30[["Time","Time_average_taxi"]].merge(df_means0[["Time","Time_average_taxi"]], left_on='Time', right_on='Time', how='outer')
df_Time_average_taxi.rename(columns={'Time_average_taxi_x': 'Time_average_taxi30', 'Time_average_taxi_y': 'Time_average_taxi0'}, inplace=True)
plot_time_dock=df_Time_average_taxi.plot(title=alternative, x='Time', y=['Time_average_taxi30','Time_average_taxi0'], xticks=np.arange(0, x, 48.0))
plot_time_dock.set_xlabel("hour of simulation")
plot_time_dock.set_ylabel("time in hours")
fig=plot_time_dock.get_figure()
fig.savefig("Results/Plots/taxi_time_"+alternative)
print(df_Time_average_taxi.mean(axis=0))

#
df_Time_average_unloading=df_means30[["Time","Time_average_unloading"]].merge(df_means0[["Time","Time_average_unloading"]], left_on='Time', right_on='Time', how='outer')
df_Time_average_unloading.rename(columns={'Time_average_unloading_x': 'Time_average_unloading30', 'Time_average_unloading_y': 'Time_average_unloading0'}, inplace=True)
plot_time_unloading=df_Time_average_unloading.plot(title=alternative, x='Time', y=['Time_average_unloading30','Time_average_unloading0'], xticks=np.arange(0, x, 48.0))
plot_time_unloading.set_xlabel("hour of simulation")
plot_time_unloading.set_ylabel("time in hours")
fig=plot_time_unloading.get_figure()
fig.savefig("Results/Plots/unloading_time_"+alternative)
print(df_Time_average_unloading.mean(axis=0))

df_Time_average_warehoused=df_means30[["Time","Time_average_warehouse"]].merge(df_means0[["Time","Time_average_warehouse"]], left_on='Time', right_on='Time', how='outer')
df_Time_average_warehoused.rename(columns={'Time_average_warehouse_x': 'Time_average_warehouse30', 'Time_average_warehouse_y': 'Time_average_warehouse0'}, inplace=True)
plot_time_warehouse=df_Time_average_warehoused.plot(title=alternative, x='Time', y=['Time_average_warehouse30','Time_average_warehouse0'], xticks=np.arange(0, x, 48.0))
plot_time_warehouse.set_xlabel("hour of simulation")
plot_time_warehouse.set_ylabel("time in hours")
fig=plot_time_warehouse.get_figure()
fig.savefig("Results/Plots/warehouse_time_"+alternative)
print(df_Time_average_warehoused.mean(axis=0))

#total time
df_Time_average_total=df_means30[["Time","average_TIS"]].merge(df_means0[["Time","average_TIS"]], left_on='Time', right_on='Time', how='outer')
df_Time_average_total.rename(columns={'average_TIS_x': 'average_TIS30', 'average_TIS_y': 'average_TIS0'}, inplace=True)
plot_time_total=df_Time_average_total.plot(title=alternative, x='Time', y=['average_TIS30','average_TIS0'], xticks=np.arange(0, x, 48.0))
plot_time_total.set_xlabel("hour of simulation")
plot_time_total.set_ylabel("time in hours")
fig=plot_time_total.get_figure()
fig.savefig("Results/Plots/average_TIS_"+alternative)
print(df_Time_average_total.mean(axis=0))
```

```python
#IDLE CARGO

#max 150cargo units idle at taxilane
df_cargo_at_taxilane=df_means30[["Time","cargo_at_taxilane"]].merge(df_means0[["Time","cargo_at_taxilane"]], left_on='Time', right_on='Time', how='outer')
df_cargo_at_taxilane.rename(columns={'cargo_at_taxilane_x': 'cargo_at_taxilane30', 'cargo_at_taxilane_y': 'cargo_at_taxilane0'}, inplace=True)
plot_idle_dock=df_cargo_at_taxilane.plot(title=alternative, x='Time', y=['cargo_at_taxilane30','cargo_at_taxilane0'], xticks=np.arange(0, x, 48.0))
plot_idle_dock.set_xlabel("hour of simulation")
plot_idle_dock.set_ylabel("cargo units")
fig=plot_idle_dock.get_figure()
fig.savefig("Results/Plots/idle_at_taxilane"+alternative)
print(df_cargo_at_taxilane.mean(axis=0))

df_Time_average_warehouse=df_means30[["Time","cargo_at_dolly_unloading"]].merge(df_means0[["Time","cargo_at_dolly_unloading"]], left_on='Time', right_on='Time', how='outer')
df_Time_average_warehouse.rename(columns={'cargo_at_dolly_unloading_x': 'cargo_at_dolly_unloading30', 'cargo_at_dolly_unloading_y': 'cargo_at_dolly_unloading0'}, inplace=True)
plot_idle_unloading=df_Time_average_warehouse.plot(title=alternative, x='Time', y=['cargo_at_dolly_unloading30','cargo_at_dolly_unloading0'], xticks=np.arange(0, x, 48.0))
plot_idle_unloading.set_xlabel("hour of simulation")
plot_idle_unloading.set_ylabel("cargo units")
fig=plot_idle_unloading.get_figure()
fig.savefig("Results/Plots/idle_at_unloading"+alternative)
print(df_Time_average_warehouse.mean(axis=0))

#max 200 cargo units idle warehouse
df_Time_average_warehoused=df_means30[["Time","cargo_at_warehouse"]].merge(df_means0[["Time","cargo_at_warehouse"]], left_on='Time', right_on='Time', how='outer')
df_Time_average_warehoused.rename(columns={'cargo_at_warehouse_x': 'cargo_at_warehouse30', 'cargo_at_warehouse_y': 'cargo_at_warehouse0'}, inplace=True)
plot_idle_warehouse=df_Time_average_warehoused.plot(title=alternative, x='Time', y=['cargo_at_warehouse30','cargo_at_warehouse0'], xticks=np.arange(0, x, 48.0))
plot_idle_warehouse.set_xlabel("hour of simulation")
plot_idle_warehouse.set_ylabel("cargo units")
fig=plot_idle_warehouse.get_figure()
fig.savefig("Results/Plots/idle_at_warehouse"+alternative)
print(df_Time_average_warehoused.mean(axis=0))
```

```python
#total idle
df_total_idle=pd.DataFrame()
df_total_idle["Time"]=df_means30["Time"]
df_total_idle["total_idle30"]=df_means30["cargo_at_taxilane"]+df_means30["cargo_at_dolly_unloading"]+df_means30["cargo_at_warehouse"]
df_total_idle["total_idle0"]=df_means0["cargo_at_taxilane"]+df_means0["cargo_at_dolly_unloading"]+df_means0["cargo_at_warehouse"]
plot_idle_total=df_total_idle.plot(title=alternative, x='Time', y=['total_idle30','total_idle0'], xticks=np.arange(0, x, 48.0))
plot_idle_total.set_xlabel("hour of simulation")
plot_idle_total.set_ylabel("cargo units")
fig=plot_idle_total.get_figure()
fig.savefig("Results/Plots/idle_total"+alternative)
print(df_Time_average_warehoused.mean(axis=0))
```
